```python
def check_win(word_so_far): #A Boolean function that returns True if all the letters in the secret word are included in the list of letters the user guessed. Otherwise, the function returns False
    if "_ " not in word_so_far and "_" not in word_so_far and " _" not in word_so_far:
        return True
    else:
        return False
```

```python
def show_hidden_word(secret_word, old_letters_guessed): #A function that returns a string which consists of letters and lower hopes. The string shows the letters from the old_letters_guessed list that are in the secret_word string in their position, and the rest of the letters in the string (which the player has not yet guessed) as underlines.
    secret_list = list(secret_word)
    guess_so_far = ""
    
    for letter in secret_list:
        if letter in old_letters_guessed:
            guess_so_far += letter + " "
        else:
            guess_so_far += "_ "
            
    return guess_so_far[ : -1] #delete the last " "
```

```python
def try_update_letter_guessed(letter_guessed, old_letters_guessed):#The function uses the check_valid_input function to know if the character is valid and has not been guessed already or the character is not valid and / or is already on the guess list.
# If the character is invalid or the character has been guessed before, the function prints the character X (as a capital letter), below it the list of letters that have already been guessed and returns false. If the character is correct and has not been guessed before - the function adds the character to the guess list and returns true.    
    character_integrity = check_valid_input(letter_guessed, old_letters_guessed)
    letter_guessed = letter_guessed.lower()
    
    if character_integrity:
        old_letters_guessed += letter_guessed
        return True
    else:
        print("❌")
        false_list_print(old_letters_guessed)
        return False
```

```python
def false_list_print(old_letters_guessed): #Prints the list of letters the user has already selected
    print("You already try this letter")
    old_letters_guessed.sort()
    print(" -> ".join(old_letters_guessed))
```

```python
def check_valid_input(letter_guessed, old_letters_guessed): #A Boolean function that receives a character and a list of letters that the user has previously guessed. The function checks two things: the correctness of the input and whether it is legal to guess this signal (i.e., the player has not guessed this signal before) and returns true or false accordingly
    letter_guessed = letter_guessed.lower()
    
    if len(letter_guessed)> 1:
        return False

    elif not(letter_guessed.isalpha()):
        return False

    elif letter_guessed in old_letters_guessed:
        return False
    
    else:
        return True
```

```python
def choose_word(file_path, index): #The function opens a text file with a list of words and the user selects a word (to guess) by selecting an index
    f = open (file_path , "r")
    read_words = f.read()
    words_data = read_words.split(" ")
    list_of_words = []
    guessing_word = ""
    
    for word in words_data:
        if word not in list_of_words:
            list_of_words.append(word)

    while True:
        if index < len(words_data):
            guessing_word = words_data[index-1]
            break
        else:
            index = index - len(words_data)
    
    return(guessing_word)
```

```python
def success_tone():
    pygame.mixer.init()
    success = r'.\sound\success.mp3'
    good_sound = pygame.mixer.Sound(success)
    good_sound.play()
```

```python
def unsuccess_tone():
    pygame.mixer.init()
    unsuccess = r'.\sound\unsuccess.mp3'
    bad_sound = pygame.mixer.Sound(unsuccess)
    bad_sound.play()
```
